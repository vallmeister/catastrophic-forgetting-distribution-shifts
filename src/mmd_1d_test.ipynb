# Testing MMD with simple numbers

```python
from metrics import mmd_rbf
import matplotlib.pyplot as plt
```

```python
time_steps = []
mmds = []
for t in range(50):
    time_steps.append(t)
    mmds.append(mmd_rbf([[1.]], [[0.1 * t]], 2))
```

```python
plt.figure(figsize=(12, 6))

plt.plot(time_steps, mmds, marker='o', linestyle='-', color='b')

plt.title('MMD in 1D')
plt.xlabel('Time Steps')
plt.ylabel('MMD')
plt.grid(True)
plt.show()
plt.close()
```
